# Speed dating 

Ovaj dataset sadrži podatke koji su prikupljeni tokom vršenja eksperimenta na univrzitetu Columbia.
Zadatak eksperimenta je bio da spoje dva učesnika različitog pola koji su nakon četvorominutnog sastanka(dejt-a) trebali da daju ocenu partneru sa kojim su se sastali i da se izjasne da li bi izašli na spoj sa tim partnerom.
Ocenjivanje se sprovodilo u dve etape:
    1. pre sastanka (demografija, zanimanja, očekivanja karakteristika partnera, ocenjivanje svojih karakteristika)
    2. posle sastanka (ocenjivanje partnera, da li bi izašli sa njim, koliko im se sviđa ...)
    
    
Atributi sa objašnjenjem:
 * gender: Pol ispitanika  
 * age: Broj godina ispitanika  
 * age_o: Broj godina partnera  
 * d_age: Razlika u godinama
 * race: Poreklo (rasa) ispitanika  
 * race_o: Poreklo (rasa) partnera  
 * samerace: Razlika u poreklu : 0-različito poreklo (rasa), 1-isto poreklo (rasa)
 * importance_same_race: Koliko je ispitaniku važno da je partner istog porekla (rase)?
 * importance_same_religion: Koliko je ispitaniku važno da je partner iste vere (religije)?
 * field: Oblast studiranja ispitanika
 

 * pref_o_attractive: Koliko je partneru važna privlačnost kod ispitanika  
 * pref_o_sinsere: Koliko je partneru važna iskrenost kod ispitanika 
 * pref_o_intelligence: Koliko je partneru važna inteligencija ispitanika
 * pref_o_funny: Koliko je partneru važna zabava(zanimljivost,interesantnost) kod ispitanika  
 * pref_o_ambitious: Koliko je partneru važna ambicioznost kod ispitanika 
 * pref_o_shared_interests: Koliko je partneru važno da ima zajednička interesovanja sa ispitanikom 


 * attractive_o: Partnerova ocena o atraktivnosti ispitanika
 * sincere_o: Partnerova ocena o iskrenosti ispitanika  
 * intelligence_o: Partnerova ocena o inteligenciji ispitanika  
 * funny_o: Partnerova ocena o zanimljivosti ispitanika  
 * ambitous_o: Partnerova ocena o ambicioznosti ispitanika 
 * shared_interests_o: Partnerova ocena o zajedničkim interesovanjima sa ispitanikom
 
 
 * attractive_important: Koliko je ispitaniku važno da je partner privlačan 
 * sincere_important: Koliko je ispitaniku važno da je partner iskren  
 * intellicence_important: Koliko je ispitaniku važno da je partner inteligentan
 * funny_important: Koliko je ispitaniku važno da je partner zabavan
 * ambtition_important: Koliko je ispitaniku važno da je partner ambiciozan
 * shared_interests_important: Koliko je ispitaniku važno da ima zajednička interesovanja sa partnerom


 * attractive: Ispitanikova ocena o svojoj atraktivnosti 
 * sincere: Ispitanikova ocena o svojoj iskrenosti 
 * intelligence: Ispitanikova ocena o svojoj inteligenciji   
 * funny: Ispitanikova ocena o svojoj zanimljivosti (zabavi)   
 * ambition: Ispitanikova ocena o svojoj ambicioznosti  


 * attractive_partner: Ispitanikova ocena o atraktivnosti partnera  
 * sincere_partner: Ispitanikova ocena o iskrenosti partnera  
 * intelligence_partner: Ispitanikova ocena o inteligenciji partnera     
 * funny_partner: Ispitanikova ocena o zanimljivosti(zabavi) partnera    
 * ambition_partner: Ispitanikova ocena o ambiciji partnera     
 * shared_interests_partner: Ispitanikova ocena o deljenim interesima sa partnerom   


Ispitanikovi lični insteresi [1-10]: 
 * sports: interesovanje prema bavljenju sportom  
 * tvsports: interesovanje prema TV sportovima  
 * exercise: vežbanje (workout treniranje, trčanje)  
 * dining: interesovanje za večeru (sa društvom u nekom restoranu/ćevapdžinici)  
 * museums: interesovanje prema muzejima
 * art: interesovanje prema umetnosti
 * hiking: interesovanje prema šetnji (voliš da prošetaš kada je napolju lep dan)
 * gaming: interesovanje prema kompjuterskim igricama 
 * clubbing: interesovanje za izlaženje po klubovima  
 * reading: interesovanje za čitanje (knjiga, časopisa...) 
 * tv: interesovanje za gledanje televizije 
 * theater: interesovanje za pozorište 
 * movies: interesovanje za filmove 
 * concerts: interesovanje za koncerte 
 * music: interesovanje za muziku 
 * shopping: interesovanje za kupovinu  
 * yoga: interesovanje za jogu  


 * interests_correlate: Korelacija između ispitanikovih i partnerovih ličnih interesa
 * expected_happy_with_sd_people: Koliko je ispitanik srećan(uzbuđen) da će upoznati partnera na speed-dating događaju ?
 * expected_num_interested_in_me: Na osnovu 20 ljudi kojih je ispitanik upoznao, koliko njih će biti zainteresovano da ide na sastanak (dejt) sa njim?  
 * expected_num_matches: Koliko ispitanik očekuje da će imati spojeva?  
 * like: Koliko se ispitaniku sviđa partner ?  
 * guess_prob_liked: Koliko ispitanik misli da će se dopasti partneru?   
 * met: Da li je ispitanik poznavao partnera od ranije 0-Nije, 1-Jeste?   
 * decision: Ispitanikova odluka za spoj  0-Ne, 1-Da
 * decision_o: Parnerova odluka za spoje  0-Ne, 1-Da
 * match: spoj 0-Ne, 1-Da
 
 

Nad ovim dataset-om kao zadatak imam da primenim algoritam klasifikiacije i moj izbor je K-Nearest-Neighbors algoritam

Funkcije za zamenu null vrednosti

```python
def replace_null_0_1(cols):
    col=cols[0]
    if pd.isna(col):
        value=randint(0,1)
        return value
    else:
        return col
```

```python
speed_dating['met']= speed_dating[['met']].apply(replace_null_0_1,axis=1)
```

```python
def change_null_with_mean_rounded_numbers(col):
    atribute = col
    
    values = speed_dating[atribute]
    atribute_mean =round(speed_dating[col].mean(skipna=True))
    length=values.size
    
    i=0
    while(i<length):
        if pd.isna(values[i]):
            values[i]=atribute_mean
        i=i+1
        
    return values
```

```python
speed_dating['guess_prob_liked']=change_null_with_mean_rounded_numbers(col='guess_prob_liked')
speed_dating['like']=change_null_with_mean_rounded_numbers(col='like')
speed_dating['ambition_partner']=change_null_with_mean_rounded_numbers(col='ambition_partner')
speed_dating['shared_interests_partner']=change_null_with_mean_rounded_numbers(col='shared_interests_partner')
speed_dating['sinsere_o']=change_null_with_mean_rounded_numbers(col='sinsere_o')
speed_dating['intelligence_o']=change_null_with_mean_rounded_numbers(col='intelligence_o')
speed_dating['funny_o']=change_null_with_mean_rounded_numbers(col='funny_o')
speed_dating['ambitous_o']=change_null_with_mean_rounded_numbers(col='ambitous_o')
```

```python
#brisanje svih redova koji sadrže null vrednost
speed_dating.dropna(axis=0,inplace=True)
```

## KNN algoritam

### Standardizacija vrednosti

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(speed_dating.drop('match',axis=1))
```

```python
scaled_features = scaler.transform(speed_dating.drop('match',axis=1))
scaled_features
```

###  Kreiranje algoritma, trening i testiranje

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,speed_dating['match'],
                                                    test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

## Unapređenje

```python
error_rate = []

for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Ocena greške vs. K vrednost')
plt.xlabel('K vrednost')
plt.ylabel('Ocena greške')
```

Ako pokrenete algoritam potrebno je promeniti vrednost atributa n_neightbors na najnižu ocenu greške

```python
knn = KNeighborsClassifier(n_neighbors=15)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
